```python
import pandas as pd
import csv
```

```python
txt_file_path = 'data/order.txt'
csv_file_path = 'data/order.csv'
```

```python
def process_line(line):
    # Split line at spaces
    parts = line.split()
    # Check if there is an array in the line
    if '[' in line:
        # Find where the array starts and ends
        start_index = line.index('[')
        end_index = line.index(']') + 1
        # Extract the array as a string
        array_string = line[start_index:end_index]
        # Remove the array from the parts
        parts = line[:start_index].split()
        # Append the array string as a single element
        parts.append(array_string)
    return parts

file_encoding = 'utf-8'  # Modify as needed

# Open the txt file with specified encoding and read lines
with open(txt_file_path, 'r', encoding=file_encoding) as file:
    lines = file.readlines()
```

```python
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)

    # Iterate over each line in the txt file
    for line in lines:
        # Process each line to handle arrays correctly
        values = process_line(line)
        print('values :::',values)
        # Write the line of values to the CSV file
        csvwriter.writerow(values)
```
